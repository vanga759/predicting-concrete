3.Checking for null and missing values .Since there are no missing values in the dataset we dont have to fill them.

```python
data.isnull().any()
```

As we saw in the scatter plot  between coarse aggregate and concrete compressive strength and fine aggregate and concrete compressive strength since both are very similar we can drop any one of them.


```python
df=data.drop("Fine Aggregate",axis=1)           
```

```python
df
```

4.Separating the dependent variables and independent variables.

```python
x=df.iloc[:,0:7].values
```

```python
x
```

```python
y=df.iloc[:,7:].values
```

```python
y
```

Since there are no categorical variables in the dataset we dont need to apply one hot encoding and label encoding.

Since the values are highly varying we can do feature scaling to standardize them.

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
y=sc.fit_transform(y)
```

```python
y
```

5.Splitting the data into train and test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_test
```

```python
x_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100,random_state=0)
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
